```python
import json
# Specify the path to your JSON file
json_file_path = r"C:\Users\PC\Downloads\german_analyst_jobs.json"

# Open and read the JSON file
with open(json_file_path, 'r') as file:
    json_data = json.load(file)

# Now, you can work with the 'json_data' variable, which contains the JSON content.
```

```python
#Viewing the German Json Data
json_data
```

```python
#Import googletranslate library
from googletrans import Translator
```

```python
# Initialize the translator
translator = Translator()

# Function to recursively translate a dictionary
def translate_dict(d):
    for key, value in d.items():
        if isinstance(value, str):
            # Translate the string from German to English
            translated_text = translator.translate(value, src='de', dest='en')
            d[key] = translated_text.text
        elif isinstance(value, dict):
            # Recursively translate nested dictionaries
            translate_dict(value)

# Translate the JSON data
translate_dict(json_data)

# Now, your 'json_data' should contain the translated values
```

```python
#View Translated json file
json_data
```

```python
import json

# Specify the path where you want to save the translated JSON data
output_file_path = r"C:\Users\PC\Downloads\translated_data.json"

# Save the translated JSON data to the file
with open(output_file_path, 'w', encoding='utf-8') as output_file:
    json.dump(json_data, output_file, ensure_ascii=False, indent=4)

# Display a message indicating the successful export
print(f'Translated data has been saved to {output_file_path}')
```
